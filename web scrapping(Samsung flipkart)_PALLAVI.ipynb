# Visualisation

```python
plt.bar(data_flipkart_sam['Price'], data_flipkart_sam['Rating'])    #pass columns here
plt.title('Price vs Rating', fontsize=16, color='r')                #pass title here
plt.xlabel('Price')
plt.ylabel('Rating')
plt.xticks(rotation=90)   # this 90 is angle
plt.grid(True)
plt.show()
```
